# **Outlier Detection**

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(20, 10))
plt.boxplot(car_prediction_df[['year','mileage','mpg','tax','price']])
plt.title('Box Plot for Target Variable')
plt.ylabel('Target')
plt.show()
```

```python
# Assuming 'car_prediction_df' is your DataFrame

def detect_outliers(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    return outliers

# Checking for outliers in each feature
for feature in ['year', 'mileage', 'mpg', 'tax', 'price']:
    outliers = detect_outliers(car_prediction_df, feature)
    print(f"Outliers in {feature}:")
    print(outliers)
```

```python
def remove_outliers(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]

# Apply to each feature
for feature in ['year', 'mileage', 'mpg', 'tax', 'price']:
    car_prediction_df = remove_outliers(car_prediction_df, feature)
```

```python
plt.figure(figsize=(20, 10))
plt.boxplot(car_prediction_df[['year','mileage','mpg','tax','price']])
plt.title('Box Plot for Target Variable')
plt.ylabel('Target')
plt.show()
```

# **Data Preprocessing**

```python
#Label Encoder
from sklearn.preprocessing import LabelEncoder,StandardScaler
le= LabelEncoder()

columns=['brand','model','transmission','fuelType','engineSize']

for i in columns:
  car_prediction_df[i]=le.fit_transform(car_prediction_df[i])
car_prediction_df
```

```python
#StandardScaler
sc= StandardScaler()

sc_columns=['year','mileage','mpg','tax']

for j in sc_columns:
  car_prediction_df[j]=sc.fit_transform(car_prediction_df[[j]])

car_prediction_df
```

```python
#Dependent Variable
X=car_prediction_df.iloc[:,:-1]
X
```

```python
#Target Variable
y=car_prediction_df.iloc[:,-1]
y
```

# **Train Test Splitting**

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lr_model=LinearRegression()

lr_model.fit(X_train,y_train)

y_pred=lr_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

# **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)


rf_model.fit(X_train, y_train)


y_pred = rf_model.predict(X_test)


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```
